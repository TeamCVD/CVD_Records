```python
import numpy as np
import pandas as pd
import os

from google.colab import drive
drive.mount('/content/drive')

os.listdir("/content/drive/My Drive/LSEG_Dataset")
```

```python
cwe119_Data = pd.read_csv("/content/drive/My Drive/LSEG_Dataset/cwe119_Data_tp.csv")
cwe120_Data = pd.read_csv("/content/drive/My Drive/LSEG_Dataset/cwe120_Data_tp.csv")
cwe469_Data = pd.read_csv("/content/drive/My Drive/LSEG_Dataset/cwe469_Data_tp.csv")
cwe476_Data = pd.read_csv("/content/drive/My Drive/LSEG_Dataset/cwe476_Data_tp.csv")
cweOTH_Data = pd.read_csv("/content/drive/My Drive/LSEG_Dataset/cweOTH_Data_tp.csv")

data = pd.read_csv("/content/drive/My Drive/LSEG_Dataset/marked_sard_vdisc_train.csv")
```

```python
cwe119_vulnerable = cwe119_Data[(cwe119_Data['CWE-119']==True)]
cwe120_vulnerable = cwe120_Data[(cwe120_Data['CWE-120']==True)]
cwe469_vulnerable = cwe469_Data[(cwe469_Data['CWE-469']==True)]
cwe476_vulnerable = cwe476_Data[(cwe476_Data['CWE-476']==True)]
cweOTH_vulnerable = cweOTH_Data[(cweOTH_Data['CWE-OTHERS']==True)]
non_vulnerable = data[(data['CWE-119']==False)&(data['CWE-120']==False)&(data['CWE-469']==False) & (data['CWE-476']==False) & (data['CWE-OTHERS']==False)]
```

```python
print(cwe119_vulnerable.shape)
print(cwe120_vulnerable.shape)
print(cwe469_vulnerable.shape)
print(cwe476_vulnerable.shape)
print(cweOTH_vulnerable.shape)
print(non_vulnerable.shape)
```

```python
non_vulnerable = non_vulnerable[:(non_vulnerable.shape[0]//100)]

print(cwe119_vulnerable.shape)
print(cwe120_vulnerable.shape)
print(cwe469_vulnerable.shape)
print(cwe476_vulnerable.shape)
print(cweOTH_vulnerable.shape)
print(non_vulnerable.shape)
```

```python
cwe119_Data = pd.concat([cwe119_vulnerable,non_vulnerable])
cwe120_Data = pd.concat([cwe120_vulnerable,non_vulnerable])
cwe469_Data = pd.concat([cwe469_vulnerable,non_vulnerable])
cwe476_Data = pd.concat([cwe476_vulnerable,non_vulnerable])
cweOTH_Data = pd.concat([cweOTH_vulnerable,non_vulnerable])
```

```python
cwe119_Data = cwe119_Data.loc[:,['code','CWE-119','Label']]
cwe120_Data = cwe120_Data.loc[:,['code','CWE-120','Label']]
cwe469_Data = cwe469_Data.loc[:,['code','CWE-469','Label']]
cwe476_Data = cwe476_Data.loc[:,['code','CWE-476','Label']]
cweOTH_Data = cweOTH_Data.loc[:,['code','CWE-OTHERS','Label']]
```

```python
cwe119_location = "/content/drive/My Drive/LSEG_Dataset/cwe119_Data_op.csv"
cwe120_location = "/content/drive/My Drive/LSEG_Dataset/cwe120_Data_op.csv"
cwe469_location = "/content/drive/My Drive/LSEG_Dataset/cwe469_Data_op.csv"
cwe476_location = "/content/drive/My Drive/LSEG_Dataset/cwe476_Data_op.csv"
cweOTH_location = "/content/drive/My Drive/LSEG_Dataset/cweOTH_Data_op.csv"

file_paths = [cwe119_location,cwe120_location,cwe469_location,cwe476_location,cweOTH_location]

for path in file_paths:
  if os.path.exists(path):
    os.remove(path)
    print(f"File '{path}' has been deleted.")

cwe119_Data.to_csv(cwe119_location,index=False)
cwe120_Data.to_csv(cwe120_location,index=False)
cwe469_Data.to_csv(cwe469_location,index=False)
cwe476_Data.to_csv(cwe476_location,index=False)
cweOTH_Data.to_csv(cweOTH_location,index=False)
```
